# SVM para Regresión

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.sort(5*np.random.rand(200,1),axis=0)
err = np.random.normal(0,0.2,200)
Y = np.sin(X).ravel()**3 + err
#Y[::5] += 3*(0.5 - np.random.rand(40))
plt.scatter(X,Y, color="darkorange", label="data")
```

```python
from sklearn.svm import SVR
```

```python
C=1e5
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)
svr_pol = SVR(kernel="poly", C=C, degree=3)
```

```python
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16,9))
plt.scatter(X,Y,color="darkorange", label ="data")
plt.plot(X,y_lin, color="blue", lw = lw, label = "SVM Lineal")
plt.plot(X,y_rbf, color="green", lw=4, label="SVM Radial")
plt.plot(X,y_pol, color="c", label="SVM Polinómico")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vector Regression")
plt.legend()
plt.show()
```
