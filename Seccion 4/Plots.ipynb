## Histogramas de frecuencias

```python
plt.hist(data["Day Calls"], bins = 20)
plt.xlabel("Numero de llamadas al día")
plt.ylabel("Frecuencia")
plt.title ("Histograma de número de llmadas al dia")
```

## Para saber cuantas diviciones a un histograma se le aplican la regla de struges
## que es c = 1 + log base 2 de M


```python
# Para la funcion de logaritmos se usa la libreria de numpy
plt.hist(data["Day Calls"], bins = int (np.ceil(1 + np.log2(3333))))
## El metodo ceil redondea hacia arriba
# se usa el int para castear el dato
plt.xlabel("Numero de llamadas al día")
plt.ylabel("Frecuencia")
plt.title ("Histograma de número de llmadas al dia")
```

# Boxplot, Diagrama de caja y bigotes
##

```python
plt.boxplot(data["Day Calls"])
plt.ylabel ("Numero de llamadas diarias")
plt.title ("Boxplot de las llamadas diarias")
```

### La caja es lo importante, esta indica la gran mayoria de los datos
### El borde inferior de la caja esta situado en el porcentil 25, el borde superior el 75
### La raya de el medio de la caja es la mediana
La resta entre el porcentil 75 y el porcentil 24 es el rango interquartilico (iqr, inter quartilic range), en este caso es 27, este es el tamaño de la caja los bigotes ( o las rayas) llegan hasta 1,5 veces el iqr

```python
data["Day Calls"].describe()
```

```python
data["Day Calls"].mode()
```

```python
IQR = data["Day Calls"].quantile(0.75) - data["Day Calls"].quantile(0.25)
IQR
```

```python
pivoteInferior = data["Day Calls"].quantile(0.25) - 1.5*IQR
pivoteSuperior = data["Day Calls"].quantile(0.75) + 1.5*IQR
print ("Pivote Superior: " + str(pivoteSuperior) + "\nPivote Inferior: " + str(pivoteInferior))
```

Todo lo que no esta en la caja o afuera de 1.5 veces el iqr se llaman datos out layer (fuera de lugar), se reprecentan con los circulos
