```python
import pandas as pd 
import os
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
mainpath = "E:/Curso de Python/Repositorio/python-ml-course/datasets/"
filepath = "ads/Advertising.csv"
data = pd.read_csv(os.path.join(mainpath,filepath))
```

```python
def getCorrP (df,var1,var2):
    df["CorrN"] = (df[var1] - np.mean(df[var1]))* (df[var2] - np.mean(df[var2]))
    corrP = sum(df["CorrN"])/(df[var1].std()*df[var2].std()*(len(df)-1))
    del df["CorrN"]
    return corrP
```

```python
getCorrP(data,"TV","Sales")
```

```python
cols = data.columns.values.tolist()
cols
```

```python
y = "Sales"
for x in cols:
    if (x != y):
        print (x + ", " + y + " : " + str(getCorrP(data,x,y)))
```
