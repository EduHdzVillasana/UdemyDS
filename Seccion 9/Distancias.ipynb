```python
from scipy.spatial import distance_matrix
import pandas as pd
import os
```

```python
mainpath = "https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets"
filepath = "movies/movies.csv"
data = pd.read_csv(os.path.join(mainpath,filepath), sep=";")
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

### No hace falta normalizar ya que todas las variables tienen el mismo rango

```python
dd1 = distance_matrix(data[movies], data[movies], p=1) # Distancia de Manhattan
dd2 = distance_matrix(data[movies], data[movies], p=2) # Distancia Euclidia
dd10 = distance_matrix(data[movies], data[movies], p=10) # Distancia de Minkowski de orden 10
```

```python
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

# Clustering jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")
Z
```

```python
plt.figure(figsize=(10,5))
plt.title("Dendrograma jerárquico para el Clustering -Ward-")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], "average")
Z
```

```python
plt.figure(figsize=(10,5))
plt.title("Dendrograma jerárquico para el Clustering -Average-")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_font_size=10.0)
plt.show()
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "complete")
Z
```

```python
plt.figure(figsize=(10,5))
plt.title("Dendrograma jerárquico para el Clustering -Complete-")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(10,5))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_font_size=10.0)
plt.show()
```
